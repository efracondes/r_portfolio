# Breast Cancer 

Breast cancer causes the death of thousands of women around the world. That is why any effort to understand and/or predict this disease is essential. In this notebook, we implement logistic regression and a support vector machine to explore the possibility of predicting whether a sample was taken from a cancerous or non-cancerous breast tumor, using logistic regression and a support vector machine.

## 1. Data

### 1.1 Brief Description of the Data

The University of Wisconsin Hospitals gathered the data used for this project. The dataset consists of information on samples taken from breast tumors; some of those tumors are cancerous. 

Some of the dataset attributes are: uniformity of the cell size, uniformity of cell shape, etc. 

We have to keep in mind that the field that tells us if a sample came from a cancerous tumor is the Class field. Class 2 corresponds to the non-cancerous tumors and Class 4 to the cancerous tumors.

First of all, let's call the following packages, load the data into a DataFrame and print the first five rows so we can take a look at it.

```r
# Package
warn = getOption("warn")
options(warn=-1)
library(dplyr)
library(tidyr)
library(e1071)
library(ggplot2)
library(caTools)
set.seed(182) 
library(rfUtilities)

# Data
attribute_names <- c('sample_id', 
                     'Clump_Thickness', 
                     'Uniformity_of_Cell Size', 
                     'Uniformity_of_Cell_Shape', 
                     'Marginal_Adhesion', 
                     'Single_Epithelial_Cell_Size', 
                     'Bare_Nuclei',
                     'Bland_Chromatin',
                     'Normal_Nucleoli',
                     'Mitoses',
                     'Class')  

cancer_data <- read.csv('breast_cancer.csv', col.names = attribute_names)
head(cancer_data, 5)
```

We can plot the classes distribution.

```r
class_dist_plot <- ggplot(cancer_data, aes(Class)) +
                geom_bar(aes(fill = factor(Class))) +
                xlab('Class')+ ylab('Cases') + ggtitle('Cancer Cases')
class_dist_plot
```

Notice that there are more negative cases of breast cancer than positive cases.

### 1.2 Cleaning the Data

Before trying to apply any ML algorithm or mathematical artifact, we must clean the data. The data description tells us that the missing values in the dataset are indicated with the character '?'. 

```r
missing_val_spots <- cancer_data == '?'
cancer_data[missing_val_spots]
```

As we can see, there are 16 missing values. In the book The Data Science Design Manual page 77, Skiena provides several methods to deal with missing values. From these options we'll just pick one. 

The method that we'll follow consists of replacing the missing values by the mean value of the attribute to which it belongs. So, let's apply this process to the missing values in every column.  

```r
for (col in 1:ncol(cancer_data)){
    if ('?' %in% cancer_data[, col]){
        # We apply the round function because all values must be integers
        mean_value <- round(mean(as.numeric(cancer_data[cancer_data[col] != '?', col])))    
        cancer_data[cancer_data[col] == '?', col] <- mean_value
        # Converts the column values into numerics
        cancer_data[, col] <- as.numeric(cancer_data[, col])
    }    
}
```

```r
cancer_data[missing_val_spots]
```

The class labels are 2 and 4. We must convert them into 0 and 1.

```r
for (row in 1:nrow(cancer_data)){
    if (cancer_data[row, 'Class'] == 2){
        cancer_data[row, 'Class'] <- 0
    }else if(cancer_data[row, 'Class'] == 4){
        cancer_data[row, 'Class'] <- 1
    }
}
```

Notice that the same numeric value replaced all the '?' characters, this may be because all the missing values belonged to the same column. The next step is to split the dataset into the training and validation sets.

```r
train_size <- floor(nrow(cancer_data) * 0.75)
train_indices <- sample(1:nrow(cancer_data), train_size)
train_data <- cancer_data[train_indices,]
valid_data <- cancer_data[-train_indices,]

#X_train <- select(train_data, -Class)
#Y_train <- select(train_data, Class)
#X_test <- select(test_data, -Class)
#Y_test <- select(test_data, Class)
```

Now that the data is clean, we can continue to apply some ML algorithms.

## 2. Logistic Regression

Let's train a logistic regression model on the training data.

```r
logreg_model <- glm(formula=Class ~ ., family = binomial(link='logit'), data = train_data)
```

Now, we can evaluate the model on the training and testing sets.

```r
prob_train <- predict(logreg_model, train_data, type = 'response')
prob_valid <- predict(logreg_model, valid_data, type = 'response')
```

The function defined below will help us to obtain the class predicted by the model for every sample.

```r
get_class <- function(probability){
    if (probability > 0.5){
        return(1)
    }else{
        return(0)
    }
}

pred_train <- sapply(prob_train, get_class)
pred_valid <- sapply(prob_valid, get_class)
```

The accuracy function will help us to get the metrics of the model's performance.

```r
train_metrics <- accuracy(pred_train, train_data$Class)
valid_metrics <- accuracy(pred_valid, valid_data$Class)
```

```r
print(paste('Accuracy in the training set:',round(train_metrics$PCC, 2)))
print('Confusion Matrix (Training Set):')
train_metrics$confusion
```

```r
print(paste('Accuracy in the validation set:',round(valid_metrics$PCC, 2)))
print('Confusion Matrix (Validation Set):')
valid_metrics$confusion
```

## 3. Support Vector Machine

Now, let's train a support vector machine model on the training data.

```r
svm_model <- svm(formula=Class ~ ., data = train_data, type = 'C-classification')
```

Again, we evaluate the model on both the training and validation sets.

```r
pred_train <- as.numeric(as.character(predict(svm_model, train_data)))
pred_valid <- as.numeric(as.character(predict(svm_model, valid_data)))
```

Below we calculate the metrics.

```r
train_metrics <- accuracy(pred_train, train_data$Class)
valid_metrics <- accuracy(pred_valid, valid_data$Class)
```

```r
And finaly we get to see the results.
```

```r
print(paste('Accuracy in the training set:',round(train_metrics$PCC, 2)))
print('Confusion Matrix (Training Set):')
train_metrics$confusion
```

```r
print(paste('Accuracy in the validation set:',round(valid_metrics$PCC, 2)))
print('Confusion Matrix (Validation Set):')
valid_metrics$confusion
```

## 4. Conclusions



We implemented a logistic regression based model and a SVM based model to predict the cancerous tumors in the UWH (University of Wisconsin Hospitals) dataset. Here are the results:

* The logistic regression based model achieved an accuracy of 97.32% on the training set and 94.29% on the validation set.

* The SVM based model achieved an accuracy of 97.71% on the training set and 96.57% on the validation set.

* Also, we'd like to stress that, although both models predicted the same number of false non-cancerous tumors, the SVM based model was able to predict correctly more cancerous cases. Thus, we can consider the SVM model to be a better model for cancerous tumors' prediction. 

* This analysis we've done here is too simplistic still. For us to make a more precise model selection, we would have to use techniques such as dividing the dataset into a training,  a validation, and a test set. Also, we could make use of cross-validation.
